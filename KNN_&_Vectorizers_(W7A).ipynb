```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2, random_state=10)
ssc=StandardScaler()
softmax_reg=LogisticRegression(multi_class='multinomial', solver='sag')
pipepa = Pipeline(steps=[("ssc",ssc),("smr", softmax_reg)])
pipepa.fit(X_train,y_train)
```

```python
y_pred = pipepa.predict(X_train)
accuracy_score(y_train ,y_pred)
```

```python
y_pred_test = pipepa.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

```python
print(classification_report(y_test, y_pred_test))
```

```python
ss=StandardScaler()
sr=SGDRegressor(random_state=10)
epoch=1
chunksize=20000
for iter in pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00246/3D_spatial_network.txt", header=None, names=['OSM_ID', 'LONGITUDE', 'LATITUDE', 'ALTITUDE'], chunksize=chunksize, iterator=True):
  ss.partial_fit(iter)
  iter = ss.transform(iter)
  X_train_p=iter[:,0:3]
  y_train_p=iter[:,3]
  sr.partial_fit(X_train_p, y_train_p)
  print("iteration: ", epoch, "intercept: ", sr.intercept_, "coeff: ", sr.coef_)
  epoch+=1
```

```python
iris = load_iris()
X3,y3=iris.data,iris.target
```

```python
X_train3,X_test3,y_train3,y_test3 = train_test_split(X3,y3, test_size=0.2, random_state=10)
norm = Normalizer()
knc = KNeighborsClassifier(n_neighbors=1)
pipe = Pipeline([('scaler', Normalizer()),('knc', KNeighborsClassifier(n_neighbors=1))])
param_grid = {'knc__n_neighbors':[2,3,4]}
gscv = GridSearchCV(pipe, param_grid=param_grid, scoring='accuracy')
gscv.fit(X_train3, y_train3)
```

```python
print(gscv.best_estimator_)
```

```python
y_pred = gscv.best_estimator_.predict(X_test3)
print(classification_report(y_test3, y_pred))
```
